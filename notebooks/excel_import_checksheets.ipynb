# Excel 检查表单导入分析

本 Notebook 用于查看 `/doc/data/250918-检查表单.xlsx` 的结构，并为后续导入 CheckSheet/CheckSheetItem 表做准备。

---

## 1. 导入所需库

```python
import pandas as pd
import os

excel_path = os.path.join('doc', 'data', '250918-检查表单.xlsx')

# 检查文件是否存在
assert os.path.exists(excel_path), f"文件不存在: {excel_path}"
```

## 2. 加载Excel文件

```python
# 读取所有sheet名
sheet_names = pd.ExcelFile(excel_path).sheet_names
print('Sheet names:', sheet_names)

# 读取第一个sheet
sheet0 = pd.read_excel(excel_path, sheet_name=0)
sheet0.head()
```

## 3. 查看Excel表结构

```python
# 查看字段、数据类型和部分数据
sheet0.info()
sheet0.columns
sheet0.head(10)
```

## 4. 导入数据到DataFrame

```python
# 若有多个sheet，可遍历全部导入
sheets = pd.read_excel(excel_path, sheet_name=None)
for name, df in sheets.items():
    print(f"Sheet: {name}, 行数: {len(df)}")
    display(df.head(5))
```
